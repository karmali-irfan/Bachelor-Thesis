```python
import matplotlib.pyplot as plt

clean = plt.imread('shirt_clean.jpg') / 255  # dividing by 255 to bring the pixel values between 0 and 1
print(clean.shape)

stained = plt.imread('shirt_cropped.jpg') / 255  # dividing by 255 to bring the pixel values between 0 and 1
print(stained.shape)
```

```python
# conversion into 2D array
clean_n = clean.reshape(clean.shape[0]*clean.shape[1], clean.shape[2])
clean_n.shape
```

```python
stained_n = stained.reshape(stained.shape[0]*stained.shape[1], stained.shape[2])
stained_n.shape
```

```python
# k-means algorithm
from sklearn.cluster import KMeans
fig, axes = plt.subplots(1, 5, figsize=(12, 6))
for i in range (1,6):
    if i > 1: 
        kmeans = KMeans(n_clusters=i-1, random_state=0).fit(clean_n)
        pic2show = kmeans.cluster_centers_[kmeans.labels_]
        cluster_pic = pic2show.reshape(clean.shape[0], clean.shape[1], clean.shape[2])
        axes[i-1].imshow(cluster_pic)
        axes[i-1].set_title(f"{i-1} cluster(s)")
    else: 
        axes[i-1].imshow(clean)
        axes[i-1].set_title(f"Clean Textile")
plt.show()

fig, axes = plt.subplots(1, 5, figsize=(12, 6))
for i in range (1,6):
    if i > 1: 
        kmeans = KMeans(n_clusters=i-1, random_state=0).fit(stained_n)
        pic2show = kmeans.cluster_centers_[kmeans.labels_]
        cluster_pic = pic2show.reshape(stained.shape[0], stained.shape[1], stained.shape[2])
        axes[i-1].imshow(cluster_pic)
        axes[i-1].set_title(f"{i-1} cluster(s)")
    else: 
        axes[i-1].imshow(stained)
        axes[i-1].set_title(f"Stained Textile")
        
plt.show()
```
