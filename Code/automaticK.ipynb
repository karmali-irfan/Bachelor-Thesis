```python
# elbow method
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
import sklearn
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
X1 = [3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8]
X2 = [5, 4, 5, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3]
plt.scatter(X1,X2)
plt.show()

data = {'X1': X1, 'X2': X2}
data_frame = pd.DataFrame(data)

Sum_of_squared_distances = []
K = range(1,10)
for num_clusters in K:
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(data_frame)
 Sum_of_squared_distances.append(kmeans.inertia_)
plt.plot(K,Sum_of_squared_distances,"bx-")
plt.xlabel("‘Values of K’") 
plt.ylabel("‘Sum of squared distances/Inertia’") 
plt.title("‘Elbow Method For Optimal k’")
plt.show()
```

```python
# Silhouette Analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
silhouette_avg = []
for num_clusters in range_n_clusters:
 # initialise kmeans
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(data_frame)
 cluster_labels = kmeans.labels_
 # silhouette score
 silhouette_avg.append(silhouette_score(data_frame, cluster_labels))

plt.plot(range_n_clusters, silhouette_avg, "bx-")    
plt.xlabel("Values of K") 
plt.ylabel("Silhouette score") 
plt.title("Silhouette analysis For Optimal k")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Load the image
image = plt.imread('img/red.jpeg') / 255  # Replace 'your_image_path.jpg' with the path to your image
image_n = image.reshape(image.shape[0]*image.shape[1], image.shape[2])
print(image_n.shape)

# Flatten the image into a 2D array of pixels
print("Flattenning the image")
h, w, d = image.shape
image_2d = image.reshape(h * w, d)

# Elbow Method
Sum_of_squared_distances = []
K = range(1, 10)
print('Elbow method...')
for num_clusters in K:
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(image_2d)
    Sum_of_squared_distances.append(kmeans.inertia_)

# Silhouette Analysis
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
silhouette_avg = []
print('Silhouette Analysis...')
for num_clusters in range_n_clusters:
    # initialise kmeans
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(image_2d)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg.append(silhouette_score(image_2d, cluster_labels))

# Plotting
print('Plotting...')
fig, axs = plt.subplots(1, 4, figsize=(15, 3))
# Plot original image
axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[0].axis('off')

# Plot elbow method
axs[1].plot(K, Sum_of_squared_distances, "bx-")
axs[1].set_xlabel("Values of K")
axs[1].set_ylabel("Sum of squared distances / Inertia")
axs[1].set_title("Elbow Method For Optimal K")

# Plot silhouette analysis
axs[2].plot(range_n_clusters, silhouette_avg, "bx-")
axs[2].set_xlabel("Number of Clusters")
axs[2].set_ylabel("Silhouette Score")
axs[2].set_title("Silhouette Analysis for Optimal k")

# kmeans
kmeans = KMeans(n_clusters=2, random_state=0).fit(image_n)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
cluster_pic = pic2show.reshape(image.shape[0], image.shape[1], image.shape[2])
axs[3].imshow(cluster_pic)
axs[3].set_title(f"kMeans - 2 cluster(s)")

plt.tight_layout()
plt.show()
```
